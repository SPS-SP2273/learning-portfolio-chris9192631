**Exercise 3 (Timing Stuff)**

```python
%%timeit -n 5000
squares = [i**2 for i in range (10)]

print(squares)
##My cell magic is not working, i tried line magic also but cannnot
##Ans: you need to specify the number of loops it can run so it wont run until the insane number of default loops
```

```python
%%timeit -n 5000

squares = list()
for i in range(10):
    squares.append(i**2)
```
